```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
df = h2o.import_file('C:\\Users\\Narayan\\Downloads\\paysim1\\PS_20174392719_1491204439457_log.csv')
df.head(10)
```

```python
df.describe()
```

```python
factorslist = ['isFraud']
df[factorslist] = df[factorslist].asfactor()
```

```python
df1 = df[df['type'] == 'TRANSFER' or df['type'] == 'CASH_OUT']
y = 'isFraud'
x = df.columns
x.remove(y)
x.remove('nameOrig')
x.remove('nameDest')
x.remove('isFlaggedFraud')
```

```python
train, test = df1.split_frame([0.8])
```

```python
aml = H2OAutoML(max_models = 5, seed = 1)
```

```python
aml.train(x = x, y = y, training_frame = train)
```

```python
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
leader_model = aml.leader
```

```python
leader_model.model_performance(test)
```

```python
leader_model.varimp_plot()
```
